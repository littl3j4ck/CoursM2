```python
text_file='/home/littlejack/Documents/git/CoursM2/Techniques avanc√©es de programmation-NOEL&BONNETON/moby.txt'
letters="ABCDEFGHIJKLMNOPQRSTUVXYZ"
lcount=dict([l,0] for l in letters)

for l in open(text_file).read():
    try :
        lcount[l.upper()] +=1
    except KeyError :
        pass
    
norm=sum(lcount.values())

fig=plt.figure()
ax=fig.add_subplot(111)
x=range(26)
ax.bar(x,[lcount[l]/norm*100 for l in letters],width=0.8, color='green',alpha=0.5,align = 'center')
ax.set_xsticks(x)
ax.setxtickslabels(letters)
ax.yaxis.grid(True)
ax.set_ylabel('Frequence des lettres en %')
plt.show()

        
```

```python
fig = plt.figure()
ax=fig.add_subplot(111)
xcities=['Boston','Detroit','Houston','Phoenix']
linestyle[{'ls':'-'},{'ls': '--'},{'ls': ':'},{ls:'-.'}]
colors=[{'color': 'k'},{'color':'b'},{'color':'r'},{'color':'g'}]

for i,city in enumerate(cities) :
    #exemple √† changer pour le chemin du fichier
    filename='C:\\DATA\\{}.tsv'.format(city.lower())
    year,pop=np.loadtxt(filename,unpack=True)
    ax.plot(year,pop,label=city,**colors[i],**linestyle[i])

ax.legend(loc='upper left')
ax.set_xlim(1800,2020)
ax.set_xlable('Year')
ax.set_ylabel('Population')
```
