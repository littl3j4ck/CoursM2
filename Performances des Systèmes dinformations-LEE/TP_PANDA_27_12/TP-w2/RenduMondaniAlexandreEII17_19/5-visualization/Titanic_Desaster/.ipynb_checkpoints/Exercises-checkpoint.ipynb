### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Titanic_Desaster/train.csv). 

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url , sep = ',', encoding='ISO-8859-1')
titanic.head(10)
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()
proportions = [males, females]
plt.pie(
    proportions,
    labels = ['Males', 'Females'],
    colors = ['blue','green'],
    explode = (0.15 , 0),
    startangle = 90,
    autopct = '%1.1f%%'
    )
plt.axis('equal')
plt.title("Sex Proportion")
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
#scatter = titanic.plot.scatter(x='Age',y='Fare',color = ['blue','red'])
#plt.title("Fare x Age")
scatter = sns.lmplot(x="Age",y="Fare",data=titanic,hue="Sex",fit_reg = False)
scatter.set(title= 'Fare x Age')
```

### Step 7. How many people survived?

```python
titanic.Survived.sum()
```

### Step 8. Create a histogram with the Fare payed

```python
ValueBin = np.arange(0,500,10)
plt.hist(titanic.Fare.sort_values(ascending=False),bins = ValueBin)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Price Histogram')
```
