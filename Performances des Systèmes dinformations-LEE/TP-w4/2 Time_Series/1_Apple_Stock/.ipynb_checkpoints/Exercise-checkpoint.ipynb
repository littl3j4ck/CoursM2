# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
```

### Step 3. Assign it to a variable apple

```python
apple = pd.read_csv(url, sep = ',' )
```

### Step 4.  Check out the type of the columns

```python
apple.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
apple['Date'] = apple['Date'].astype('datetime64[ns]')
```

### Step 6.  Set the date as the index

```python
apple = apple.set_index('Date')
```

### Step 7.  Is there any duplicate dates?

```python
apple.index.duplicated()
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
apple = apple.sort_values(by=['Date'])
apple
```

### Step 9. Get the last business day of each month

```python
month = apple.resample('M').mean()
month
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
diff = apple.index.max() - apple.index.min()
diff.days
```

### Step 11.  How many months in the data we have?

```python
len(month.index)
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
fig_apple = apple['Adj Close'].plot()
fig_apple.set_title("Apple Stock")
fig = fig_apple.get_figure()
fig.set_size_inches(13.5,9)
```
