```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
#Recreate the scatter plot of b vs a
df3.plot.scatter(x='a',y='b',color = 'red',s= 45,figsize = (15,3))
plt.axis([-0.2,1.2,-0.2,1.2])
plt.show()
```

```python
#create a histogram of the 'a' column
df3['a'].plot.hist(color='blue')
plt.axis([0,1,0,70])
```

```python
#use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it
plt.style.use("ggplot")
```

```python
df3['a'].plot.hist(bins=25)
```

```python
#create a boxplot comparing the a and b columns
df3[['a','b']].plot.box()
```

```python
#create a kdeplot of the 'd' column
#figure out how to increase the linewidth and make the linestyle dashed.
df3['d'].plot.kde(lw=4,ls='--')
```

```python
#create an area plot of all the columns for just the rows up to 30
df3.loc[:30].plot.area()
```

```python
#display the legend outside of the plot.
df3.loc[:30].plot.area()
plt.legend(loc='upper right',bbox_to_anchor=(1.2,1))
plt.show()
```
