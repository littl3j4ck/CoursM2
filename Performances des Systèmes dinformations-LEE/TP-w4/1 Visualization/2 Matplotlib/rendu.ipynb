```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange (0,100)
y = x*2
z= x**2
```

```python
#import matplotlib.pyplot as plt and set %matplotlib inline. What command do you use if you aren't using the jupyter notebook?
%matplotlib inline
#or,if we don't use jupyter:
#get_ipython().magic('matplotlib inline')
```

```python
#1)create a figure object called fig using plt.figure()
#2)use add_axes to add an axis to the figure canvas at [0.0.1.1]. Call this new axis ax.
#3)plot (x,y) on that axes and set the labels and titles to match the plot below
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titre')
plt.show()
```

```python
#create a figure object and put two aces on it,ax1 and ax2. located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
#now plot (x,y) on both axes. and call your figure object to show it
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax1.set_title('titreAX1')
ax2.set_title('titreAX2')
```

```python
#create the plot below by adding two aces to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
#now use x,y and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot.
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(y,z)
ax1.set_xlabel('y')
ax1.set_ylabel('z')
ax1.set_title('titre')


ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.axis([20,22,30,50])
plt.show()
```

```python
#use plt.subplots(nrows=1,ncols=2) to create the plot below
plt.subplots(nrows=1,ncols=2)
plt.show()
```

```python
#now plot (x,y) and (x,y) on the axes. Play around with the linewidth and style
f, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'c--')
ax2.plot(x,z,lw='5')
plt.show()
```

```python
#see if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code
f, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize =(10,2))

ax1.plot(x,y,'c--')
ax2.plot(x,z,lw='5')
plt.show()
```
